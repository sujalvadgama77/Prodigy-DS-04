```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from wordcloud import WordCloud
import nltk

# Apply sentiment analysis
data['sentiment_score'] = data['im getting on borderlands and i will murder you all ,'].apply(get_sentiment_score)


# Download VADER lexicon for sentiment analysis
nltk.download('vader_lexicon')

# Load the dataset
file_path = r'C:\Users\mrvad\Desktop\task01\archive\twitter_training.csv'  # Replace with the path to your downloaded CSV file
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
print(data.head())

# Display the summary statistics of the dataset
print(data.describe())

# Display information about the dataset
print(data.info())

# Perform basic data cleaning
# Check for missing values
print(data.isnull().sum())

# Drop rows with missing values if any
data.dropna(inplace=True)

# Sentiment Analysis using VADER
sid = SentimentIntensityAnalyzer()

# Function to analyze sentiment
def get_sentiment_score(text):
    return sid.polarity_scores(text)['compound']

# Apply sentiment analysis
data['sentiment_score'] = data['text'].apply(get_sentiment_score)

# Classify sentiment as positive, negative, or neutral
def classify_sentiment(score):
    if score > 0:
        return 'positive'
    elif score < 0:
        return 'negative'
    else:
        return 'neutral'

data['sentiment'] = data['sentiment_score'].apply(classify_sentiment)

# Display the sentiment distribution
print(data['sentiment'].value_counts())

# Visualize sentiment distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='sentiment', data=data, palette='viridis')
plt.title('Sentiment Distribution')
plt.show()

# Word Cloud for Positive Sentiment
positive_text = ' '.join(data[data['sentiment'] == 'positive']['text'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(positive_text)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Positive Sentiment')
plt.show()

# Word Cloud for Negative Sentiment
negative_text = ' '.join(data[data['sentiment'] == 'negative']['text'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(negative_text)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Negative Sentiment')
plt.show()

# Word Cloud for Neutral Sentiment
neutral_text = ' '.join(data[data['sentiment'] == 'neutral']['text'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(neutral_text)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Neutral Sentiment')
plt.show()
```
